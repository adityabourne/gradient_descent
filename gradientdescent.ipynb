```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
def readdata(filename):
    df=pd.read_csv(filename)
    return df.values
x=readdata('linearx.csv')

x=x.reshape(99,)
y=readdata('lineary.csv')
y=y.reshape(99,)

print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def hypothesis(theta,x):
    return theta[0] + theta[1]*x
    
    
def error(x,y,theta):    
    m=x.shape[0]
    error_count=0.0
    
    for i in range (m):
        error_count+=(hypothesis(theta,x[i])-y[i])**2
    return error_count
                      
def gradient(x,y,theta):
    m=x.shape[0]
    grad=np.array([0.0,0.0])                  
    for i in range (m):                  
        grad[0]  =grad[0]+(hypothesis(theta,x[i])-y[i])
        grad[1]  =grad[1]+((hypothesis(theta,x[i])-y[i])*x[i])  
        
    return grad*0.5                
    
def gradientdescent(x,y,learning_rate,maxitr):
    theta=np.array([0.0,0.0])
 
    error_list=[]
    theta_list=[]
   
    for i in range(maxitr):
         grad=gradient(x,y,theta)
         e=error(x,y,theta)
         error_list.append(e)
         theta_list.append((theta[0],theta[1]))   
         grad=gradient(x,y,theta)
         theta[0] = theta[0]-learning_rate*grad[0]
         theta[1] = theta[1]-learning_rate*grad[1]
          
    return theta,error_list,theta_list                
                      
new_theta,error_list,theta_list=gradientdescent(x,y,learning_rate=0.001,maxitr=100)
print(new_theta)
plt.plot(error_list)
test_x=np.linspace(-2,4,8)
plt.show()
plt.plot(test_x,hypothesis(new_theta,test_x),color='orange')                      
                      
plt.scatter(x,y,color='red') 
plt.show()
```
